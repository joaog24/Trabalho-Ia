```python
# ðŸ”§ InstalaÃ§Ã£o e importaÃ§Ã£o de bibliotecas
!pip install seaborn --quiet

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
#1.Carregar o dataset do Spotify
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("dataset.csv")
df.head()
```

```python

# Remover colunas irrelevantes
df_clean = df.drop(columns=['Unnamed: 0', 'track_id', 'artists', 'album_name', 'track_name'])

# Selecionar os 10 gÃªneros mais comuns
top_10_genres = df_clean['track_genre'].value_counts().head(10).index
df_top10 = df_clean[df_clean['track_genre'].isin(top_10_genres)]

# Balancear: 500 amostras por classe
df_balanced = df_top10.groupby('track_genre').sample(n=500, random_state=42)

# Separar atributos e alvo
X = df_balanced.drop(columns=['track_genre'])
y = df_balanced['track_genre']
```

```python
#3. DivisÃ£o e normalizaÃ§Ã£o dos dados
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#4. Treinamento e avaliaÃ§Ã£o dos modelos
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(random_state=42),
    "Naive Bayes": GaussianNB()
}

results = {}

for name, model in models.items():
    if name in ["SVM", "Naive Bayes"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
    macro_f1 = report["macro avg"]["f1-score"]
    results[name] = {"Accuracy": acc, "Macro F1-Score": macro_f1}
```

```python
#5. ComparaÃ§Ã£o dos resultados
results_df = pd.DataFrame(results).T.reset_index().rename(columns={"index": "Algorithm"})
display(results_df)

# GrÃ¡fico de acurÃ¡cia
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="Algorithm", y="Accuracy", palette="viridis")
plt.title("ComparaÃ§Ã£o de AcurÃ¡cia entre Algoritmos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
